```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('/kaggle/input/competitions/playground-series-s6e3/train.csv')
test = pd.read_csv('/kaggle/input/competitions/playground-series-s6e3/test.csv')
```

```python
train.drop(['id'], axis = 1, inplace = True)
test_ids = test['id'].copy()
test.drop(['id'], axis = 1, inplace = True)
```

```python
train['TotalCharges'] = pd.to_numeric(train['TotalCharges'], errors='coerce').fillna(0).astype(int)
test['TotalCharges'] = pd.to_numeric(test['TotalCharges'], errors='coerce').fillna(0).astype(int)
train['MonthlyCharges'] = pd.to_numeric(train['MonthlyCharges'], errors='coerce').fillna(0).astype(int)
test['MonthlyCharges'] = pd.to_numeric(test['MonthlyCharges'], errors='coerce').fillna(0).astype(int)
```

```python
cat_cols = train.select_dtypes(include=['object']).columns
print("Categorical columns:", cat_cols)
```

```python
x = train.drop('Churn', axis = 1)
y = train['Churn']
x_test = test.copy()
x_test
```

```python
cat_cols = x.select_dtypes(include=['object']).columns
cat_cols
```

```python
from sklearn.preprocessing import LabelEncoder

for col in cat_cols:
    le = LabelEncoder()
    x[col] = le.fit_transform(x[col])
    x_test[col] = le.transform(x_test[col])



x_test
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
st = StandardScaler()

X_train_scaled = st.fit_transform(X_train)  
X_val_scaled = st.transform(X_val)           
x_test_scaled = st.transform(x_test) 
y = le.fit_transform(y)
y
```

```python
model = LogisticRegression(max_iter = 1000)

model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_val_scaled)
accuracy = accuracy_score(y_val, y_pred)
accuracy
```

```python
print(classification_report(y_val,y_pred))
```

```python

y_test_pred = model.predict_proba(x_test_scaled)[:, 1]
y_test_pred
```

```python
from sklearn.model_selection import cross_val_score
```

```python
x_scaled = st.fit_transform(x)
x_scaled
```

```python
score = cross_val_score(LogisticRegression() , x_scaled, y, cv = 7)
score
```

```python
print(score.mean())
```

```python
submission = pd.DataFrame({
    'id': test_ids,
    'Churn': y_test_pred.round(5)
})
```
